那么究竟什么是神经网络, 我们通过一个简单的例子来说明, 或许能方便理解.

假设我们有6组房屋信息的数据, 由房屋的面积和房价组成. 现在我们想找到一个能根据房屋面积预测房价的函数. 如果我们将这6组数据画在坐标系上, 将得到如下的样子:

![housing](assets/workshop0/housingprice.svg)

通过观察这些数据点, 又知道房价不可能为负数, 凭我们的直觉, 我们大概可以画出这么一条线来拟合这些点.

![housingline](assets/workshop0/housingpriceline.svg)

而这条线就是一个最简单的神经网络了, 用下面这个图画出来是这样的:

![housingpricenn](assets/workshop0/housingpricenn.svg)

我们将房屋面积作为神经网络的输入x, 通过中间这个节点输出价格, 用y表示. 而中间这个小圆圈就是一个单独的神经元, 是它实现了我们前面所说的那条直线的函数, 具体来说就是拿到输入的面积, 完成一个简单的非线性运算: 
$$
y=f(x)=max\{ 0, kx+b \}
$$

以上就是一个规模很小的神经网络, 只有一个神经元. 这个函数其实有一个名称叫做ReLU (rectified linear unit) 修正线性单元, 在这里了解就好, 暂时不用去纠结它. 

而大一点的神经网络其实就是多个神经元堆叠在一起形成的, 我们再来看一个例子.

假设我们不仅仅用房屋的面积来预测房价, 现在我们还有了房屋的其他特征, 比如卧室的数量. 你可能会想到一个很重要的影响房价的因素是可容纳的家庭成员数, 而这是与房屋总面积和卧室个数相关的; 再比如房屋所在的区域, 它又决定了住在这里的方便程度诸如附近的便利店多不多等, 还决定了这个片区是否安全, 是否是学区房啊等等条件. 那么我们可以将这一切画成这样, 每个圆圈就是我们刚才提到的那个ReLU函数. 所以基于面积和卧室我们可以评估可容纳人数, 基于所在区域可以评估方便程度, 安全程度, 教育条件; 最后又基于这些来评估房价, 因为这些是人们在评估房价时关心的和他们生活息息相关的因素. 

![housingnn2](assets/workshop0/housingnn2.svg)

在这个例子中, $x$是左边这三个输入, $y$是我们希望预测的房价, 我们把这些独立的神经元堆叠起来, 我们就有了一个规模稍大一点的神经网络. 

实际上, 上述例子中神经网络的完整结构是这样的:

![housingnn3](assets/workshop0/housingnn3.svg)

这是一个有3个输入的神经网络, 每个输入是房屋的一种特征. 给出这些特征, 神经网络的工作就是预测对应的价格. 我们注意中间这些小圆圈, 在神经网络中它们被称为"**隐藏单元**", 它们每一个都同时接收来自于$x$的3个特征. 也就是说, 与其规定第一个节点就用来表示可容纳的家庭成员数, 或者家庭成员数仅取决于特征$x_1$和$x_2$; 相反地, 我们告诉这个神经网络: 我们现在告诉你这个房屋的这些特征, 随便你怎么计算, 你自己决定这个节点表示什么. 有了多个节点后, 神经网络就有了一层层的节点, 我们将$x$这一层叫做输入层, 中间这一层叫做隐藏层. 后面的$y$叫做输出层. 神经网络的神奇之处是当你实现它以后, 你要做的只是给出若干组$(x, y)$对作为数据集, 神经网络将会自己搞清楚中间的过程是怎么回事, 而这个过程就叫做训练. 当你给的样本足够大时, 神经网络能非常出色地拟合出中间的这些各种函数, 来找出$x$到$y$的映射. 

总结一下, 神经网络其实就是由一层层的节点所组成的, 用来找出从$x$到$y$的映射规则的算法, 你不需要去人为的定义中间任何函数的参数, 只需要给它喂足够多的$(x, y)$样本数据就够了. 深度神经网络就是层数很多的神经网络, 其实没什么严格的定义, 一般认为层数≥2的就叫深度神经网络, 用深度神经网络来学习规则的所有应用/研究都叫深度学习.

4. 英文文本 -> 中文文本

## 3. 房价预测

我们将就上面的例子来实战演示一下用深度学习做个简单的房价预测. 首先我们下载需要用到的数据

链接: https://pan.baidu.com/s/1b6932ahxmJ9qAERt6lx5Lw 

提取码: **fe3c**

数据主要包括2014年5月至2015年5月美国King County的房屋销售价格以及房屋的基本信息。

其中训练数据主要包括10000条记录，14个字段，主要字段说明如下：

第一列“销售日期”：2014年5月到2015年5月房屋出售时的日期

第二列“销售价格”：房屋交易价格，单位为美元，是目标预测值

第三列“卧室数”：房屋中的卧室数目

第四列“浴室数”：房屋中的浴室数目

第五列“房屋面积”：房屋里的生活面积

第六列“停车面积”：停车坪的面积

第七列“楼层数”：房屋的楼层数

第八列“房屋评分”：King County房屋评分系统对房屋的总体评分

第九列“建筑面积”：除了地下室之外的房屋建筑面积

第十列“地下室面积”：地下室的面积

第十一列“建筑年份”：房屋建成的年份

第十二列“修复年份”：房屋上次修复的年份

第十三列"纬度"：房屋所在纬度

第十四列“经度”：房屋所在经度

我们试着用python来读取数据并显示其前5行:

```python
# 引入读取csv文件的库类
import pandas as pd
from IPython.display import display
from matplotlib import pyplot as plt
%matplotlib inline

headers = ["销售日期", "销售价格", "卧室数", "浴室数", "房屋面积", '停车面积', "楼层数", "房屋评分", "建筑面积", "地下室面积", "建筑年份", "修复年份", '纬度', '经度']
data = pd.read_csv("~/datasets/housing/kc_train.csv", header=None, names=headers, )
display(data.head(5))
```

我们将其分割成训练集和测试集, 训练集用来训练我们的深度学习模型(即神经网络), 而测试集中的数据不参与模型的训练, 只用来"考察"我们的模型表现如何. 做个简单的类比, 训练集就如同我们我们念高中时做的练习题, 而测试集好比是考试题. 一个好的模型是在做练习题时学到知识进而在考试中举一反三, 取得好成绩; 而非将练习题一股脑地背下来, 期望在考试的时候遇到原题直接得分. 这也是测试集数据不参与训练的原因.

```python
print("数据总量: {}".format(len(data)))

# 我们将数据集以9比1的比例分割
test_ratio = 0.1
test_size = int(len(data) * test_ratio)
train_size = len(data) - test_size

train_data = data[:train_size]
test_data = data[train_size:]
print("分割后数据大小: 训练 {} | 测试 {}".format(len(train_data), len(test_data)))
display(train_data.head(5))
display(test_data.head(5))
```

在房价预测这个任务中, 房价是我们要求得的结果, 也就是$y$, 而其他的特征都要作为输入$x$. 假设我们已经有一个训练好的深度学习模型, 我们期望怎么来使用它? 很显然是告诉模型一套房屋的若干特征$x$, 让它输出房价$y$. 所以我们需要简单地对原始的数据进行一些处理, 将它分成输入和输出, 也就是将"销售价格"这一列提出来作为输出, 剩下的列作为输入. 

```python
def input_output_split(dataset):
    y = dataset[['销售价格']]
    x = dataset.iloc[:, dataset.columns != '销售价格']
    return x, y

train_x, train_y = input_output_split(train_data)
test_x, test_y = input_output_split(test_data)

print(train_x.shape, train_y.shape)

print("显示输入数据:")
display(train_x.head(5))
print("显示输出数据:")
display(train_y.head(5))
```

在开始训练之前, 我们一般会对数据做一些预处理, 一个通常的做法是**归一化(Normalization)**, 目前我们不用去理解为什么要这么做, 以后我们会专门讲解. 另外, 我们对销售价格也做一个统一的处理, 也就是所有的价格数值除以一万, 这样做一方面方便我们查看, 另一方面的好处我们一会儿会进行简单的介绍.

```python
def norm(x):
    return (x - x.mean()) / x.std()

train_x = norm(train_x)
test_x = norm(test_x)

train_y = train_y / 10000
test_y = test_y / 10000

display(train_x.head(5))
display(train_y.head(5))

train_x, train_y = train_x.values, train_y.values
test_x, test_y = test_x.values, test_y.values
```

接下来我们用Keras来搭建类似刚才的神经网络模型. 值得注意的是, 刚才的例子中我们对于一套房的特征只有3个, 而现在我们有了13个特征, 所以我们的神经网络需要接收13个输入信号.

```python
from keras import Sequential, optimizers
from keras.layers import Dense

model = Sequential()
model.add(Dense(20, input_shape=(13,), activation='relu'))
model.add(Dense(1, activation='relu'))

opt = optimizers.adam(lr=.001)
model.compile(optimizer=opt, loss='mse')
```

在上面的代码中, 我们创建了类似于上面例子中的神经网络, 区别在于接收了13个输入信号, 并且中间的隐藏层有20个神经元. 

![nn20](assets/workshop0/nn20.svg)

下面我们来试着训练这个神经网络, 看看它的表现如何.

```python
H = model.fit(train_x, train_y, epochs=100, batch_size=16, validation_data=(test_x, test_y), verbose=0)
print("训练完成!!")
losses = H.history['loss']
plt.plot(losses)
```

在训练的过程中, 我们来稍微解释一下上面的代码中最后两行的意思. 

~~~python
opt = optimizers.adam(lr=.001)
~~~
这段代码指明了我们的优化方法叫做Adam, 同样地, 我们暂时不用理解内部的原理, 只需要将它理解为控制神经网络如何去学习的方法即可. 

~~~python
model.compile(optimizer=opt, loss='mse')
~~~
而最后一行代码除了将Adam学习方法加载到我们的神经网络模型上去之外, 还给模型设定了一个学习目标, 叫做mse, 这是mean squared error的缩写, 他的意思其实很好理解. 我们这么来看:

当我们的模型在面对第一组训练数据时, 它做的事情是拿到所有的输入$x$的值, 并且进行如下的运算:

$
\begin{align}
h_1&=k_{1,1}x_1 + k_{1,2}x_2 + ... + k_{1,13}x_{13} + b_1 \\
h_2&=k_{2,1}x_1 + k_{2,2}x_2 + ... + k_{2,13}x_{13} + b_2 \\
&...\\
h_{20}&=k_{20,1}x_1 + k_{20,2}x_2 + ... + k_{20,13}x_{13} + b_{20} \\
\end{align}
$

再应用ReLU函数得到中间影藏层的20个神经元的值:

$a_i=max\{0, h_i\}$

接下来再用隐藏层$a_i$来计算出$y$的值:

$
y=ReLU(p_1a_1 + p_2a_2 +...+ p_{20}a_{20} + c)
$

一开始, 中间所有的参数$k, b, p, c$都是随机的, 所以我们会得到一个很奇怪的$y$作为神经网络预测的这套房屋的价格, 很显然它会错得很离谱, 那我们就需要有一种标准来告诉神经网络说: 你错了. 这个标准就是

$E=(y-y_{true})^2$

它的含义是当前神经网络预测出来的结果与实际结果的差值的平方, 这个值越大就说明神经网络错得越厉害. 

神经网络这时候拿到这个值之后呢就会使用刚才我们告诉它的Adam方法去调整那些参数$k,b,p,c$, 让它针对所有的训练数据所作出的预测都能离真实的房价相差无几.

刚才我们提到过将价格除以一万, 他的另外一个用处其实就在这里. 

第一, 原来的数值会造成$E$非常大, 计算机在处理大数字上可能会出现问题;

第二, 会阻碍神经网络的学习效率;

下面我们来看看神经网络在测试集上的表现如何. 我们取测试集中的100套房屋, 将它们的价格绘制出来, 再将神经网络预测的这些房屋的价格也绘制出来, 看看到底预测得准不准:

```python
sample_houses = test_x[:100]
sample_prices = test_y[:100]

preds = model.predict(sample_houses)

# 将价格恢复为不以万为单位
predicted_prices = preds * 10000
sample_prices = sample_prices * 10000
```

```python
plt.figure(figsize=(24, 9))
plt.plot(predicted_prices, label='predicted prices')
plt.plot(sample_prices, label='true prices')
plt.ylim(bottom=0)
plt.legend()

diff = abs(sample_prices - predicted_prices)
count = 0
for i, d in enumerate(diff):
    d = d[0]
    if d <= 100000:
        continue
    plt.text(i, max(sample_prices[i], predicted_prices[i]), "{:,.2f}".format(d))
    count += 1

print("这100套房屋预测房价与真实房价差距总金额: ${:,.2f}".format(diff.sum()))
print("其中有{}组结果差距大于10万美金".format(count))
```

从这个图形我们可以看到预测出的价格跟真实的价格走向挺接近, 这个神经网络基本上把大方向都把握住了的. 但其实如果要用它来应用的话还不行, 因为误差还是很大. 在这100组实验中, 我们将所有的预测价格和实际价格相减再取绝对值得到误差, 所有误差之和能有接近1000万美元. 平均每套房的误差是10万美金. 而且这其中还有这么多组实验的差距是大于10万的, 当中不乏远大于10万误差的例子. 

但我们注意到这个神经网络一共有1个隐藏层21个神经元, 这是个非常简陋的神经网络. 刚才我们提到, 深度神经网络是指隐藏层≥2的神经网络, 那么我们看看用深度学习能不能解决这个问题:

```python
from keras.layers import Dropout, regularizers

model2 = Sequential()
model2.add(Dense(50, input_shape=(13,), activation='relu'))
model2.add(Dense(100, activation='relu'))
model2.add(Dropout(.3))
model2.add(Dense(60, activation='relu'))
model2.add(Dropout(.3))
model2.add(Dense(1, activation='relu', kernel_regularizer=regularizers.l2()))

opt = optimizers.adam(lr=.001)
model2.compile(optimizer=opt, loss='mse')
H1 = model2.fit(train_x, train_y, epochs=100, batch_size=16, validation_data=(test_x, test_y), verbose=0)
print("训练完成!!!")
plt.plot(H1.history['loss'])
```

相比于刚才的浅层神经网络, 我们这次添加了两个隐藏层, 分别有100和60个神经元, 并且还在中间使用了dropout和l2 regularization, 这些都是在深度学习中常用的防止神经网络"背答案"的措施. 如果各位有兴趣了解, 我们将在以后的活动中专门介绍.
